# Module 2: Exploration data analysis of visualization and training a model by given attributes

```python
#import library packages
import pandas as p
import matplotlib.pyplot as plt
import seaborn as s
import numpy as n
```

```python
data = p.read_excel("WSND.xlsx")
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
df=data.dropna()
```

```python
df.columns
```

```python
from sklearn.preprocessing import LabelEncoder
var_mod = ['Attack_Type']
le = LabelEncoder()
for i in var_mod:
    df[i] = le.fit_transform(df[i]).astype(int)
```

```python
df['JOIN_R'].hist(figsize=(5,5), color='brown', alpha=1)
plt.xlabel('JOIN_R')
plt.ylabel('SCH_S')
plt.title('JOIN_R & SCH_S')
```

```python
plt.boxplot(df['DATA_R'])
plt.show()
```

```python
import seaborn as s
s.boxplot(df['Attack_Type'], color='m')
```

```python
fig, ax = plt.subplots(figsize=(8,5))
ax.scatter(df['Time'],df['Consumed_Energy'])
ax.set_xlabel('Time')
ax.set_ylabel('Consumed_Energy')
plt.show()
```

```python
#Propagation by variable
def PropByVar(df, variable):
    dataframe_pie = df[variable].value_counts()
    ax = dataframe_pie.plot.pie(figsize=(8,8), autopct='%1.2f%%', fontsize = 12)
    ax.set_title(variable + ' \n', fontsize = 15)
    return n.round(dataframe_pie/df.shape[0]*100,2)

PropByVar(df, 'Attack_Type')
```

```python
#Propagation by variable
def PropByVar(df, variable):
    dataframe_pie = df[variable].value_counts()
    ax = dataframe_pie.plot.pie(figsize=(8,8), autopct='%1.2f%%', fontsize = 12)
    ax.set_title(variable + ' \n', fontsize = 15)
    return n.round(dataframe_pie/df.shape[0]*100,2)

PropByVar(df, 'ADV_R')
```

```python
# Heatmap plot diagram
fig, ax = plt.subplots(figsize=(20,8))
s.heatmap(df.corr(), ax=ax, annot=True)
```
